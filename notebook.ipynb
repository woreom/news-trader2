```python
import os
import pytz
import pandas as pd
from datetime import datetime, timedelta
import investpy
from get_data import convert_to_gmt, fix_dataframe

def get_calendar(from_date, timezone):
    
    now = datetime.now(timezone)
    gmt_format = convert_to_gmt(now)

    countries = ["United States", "Australia","Japan", "Euro Zone", "Germany", 
        "United Kingdom", "New Zealand", "Canada", "Switzerland"]

    df = investpy.news.economic_calendar(time_zone=gmt_format, countries=countries,
                                        importances=["high", "medium", "low"],
                                        from_date=(from_date - timedelta(days=1)).strftime("%d/%m/%Y"),
                                        to_date=(now + timedelta(days=1)).strftime("%d/%m/%Y"),
                                        )
    
    df = fix_dataframe(tf=df)
    # print(df)
    # df.to_excel(f'static/{now.strftime("%Y-%m-%d_%H")}.xlsx',index=False)
    return df

timezone = pytz.timezone('Asia/Tehran')
df = pd.read_excel('./static/Report_mt5.xlsx')
earliest_time = datetime.strptime(min(df['date_time']), "%Y-%m-%d %H:%M:%S")
cal = get_calendar(earliest_time, timezone)
```
